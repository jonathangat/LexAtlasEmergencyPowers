# Tables to Markdown

```python
# import libraries
import pandas as pd
import datetime
```

```python
# Function to convert datetime objects to string
def convert_datetime(cell):
    if isinstance(cell, datetime.datetime):
        return cell.strftime('%d/%m/%Y')
    return cell
```

## Emergency Powers dataset

```python
emergency_powers_dataset = pd.read_excel(r'D:\FreelanceProjects\Kings\lex_atlas\20230627_emergency_powers_datasets\Emergency_powers_dataset_full_1.xlsx')
```

```python
# Apply the function to each cell
emergency_powers_dataset.applymap(lambda cell: convert_datetime(cell)).to_html('Emergency_Powers_dataset.html', index=False, na_rep='')
```
